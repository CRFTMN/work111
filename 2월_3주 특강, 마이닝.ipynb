# 데이터 마이닝 개념정리

## 1. 데이터 마이닝 개념
  
특정한 데이터들로부터 의미있는 지식(정보)을 알아내는 것.

예시) 추천제도: 제품구매시 고객데이터 남음. 이를 이용해 추천. 
    구매상품에 대한 분석. 다음에 고객이 무엇을 구매하게 될지 유추
    
정확한 정의: 데이터 집합으로부터 유용한 정보를 추출하는 것.
        데이터 웨어하우스에서 존재하는 데이터 집합. 의미있는 지식을 발견하기위해 자동화 되넉나 반자동화된 도구를 이용하여 대량의 데이터를 탐색하고 분석하는 과정이 데이터 마이닝.
        
        더욱 상세하게는 패턴인식 기술들을 이용하여 데이터 저장소에 저장된 데이터를 조사함. 새로운 상관관계 패턴 추세등을 발견
        


고객에 적합한 정보를 주는 마케팅을 위해 데이터 마이닝 기술이 필수적

__데이터 마이닝은 통계학과 기계학습을 합쳐 놓은 기술__

전통통계는 나름대로의 지식을 가지고 가정을 만들어서 이 가정이 맞는지를 분석
(어떻게 보면 편견과 상식을 가지고_개연적인 접근)

데이터 마이닝은 현실적인 데이터에 대한 _가정이 없고_ 미래를 예측하는 데 주목적을 두는 기술이다. 데이터만 가지고 새로운 지식을 생성한 후,지식적용 과정 필요  
알고리즘이 비선형성에 기반, 모형에 대해 Robust(상치/에러값으로 부터 영향을 크게 받지 않는 통계량)한 결과

이 말은 데이터 마이닝은 지식의 생성과정이므로 차후의 액션플랜과정이 따로 필요하다는 의미. 2번의 단계를 거쳐야 정말 목적적합한 수준의 분석내용이 나온다는 것.


※ Robust 
> - 사전적정의_건장한 
> - 통계적정의_상치/에러값으로 부터 영향을 크게 받지 않는 (건장한) 통계량 /데이터의 특성을 보다 건강하게 반영한.
> - 머신러닝정의_범용적인 데이터도 적합한 모델을 의미. generalization에 유용.

## 3. 데이터 마이닝 분석기법

다양한 분석 기법들이 존재  

1. 연관 규칙분석  
  
- Association Rules  
Association 규칙_ 데규모의 데이터 항목들 중에서 유용한 연관성과 상관관계를 찾는 기법  
__상품, 서비스 간__ 의 관계를 살펴보고 유용한 규칙 찾기(상품 조합을 봄으로_장바구니 분석이라고도 call)  

- Sequential Pattern Rules      
사건이나 거래의 순서를 갖고 일어나는 규칙을 발견하는 분석(순서에 포커스)  
고객의 움직임을 고려하는 방식(ex.고객이 어떠한 순서로 장바구니에 물건담음?)  

2. 군집 분석  
사전 정보가 없는 데이터에 주어진 관측값을 이용 유사한 집단으로 그룹화하여 각  집단의 성격을 파악하기 기법  
모집단을 정의 되지않은 부분집합으로 분류  
클러스터링이 끝난후에 특성 파악가능_유사도측정, 거리함수, 군집분석  
계층적 방법 : K-means / 비계층적 방법: 병합적인 분할적인  


3. 분류  
그룹의 수와 특성이 정해져있는 데이터의 특성을 파악하여 서로다른 집단으로 구분  
의사결정나무가 대표적  


4. 인공신경망_딥러닝   
인간이 의사결정방식을 통해 학습하여 결과를 도출하는 방법  
뇌신경망 원리이용  
node들의 연결. Weight로 연결_학습과정은 Weight를 조절하는 과정  
  

5. 사레기반 추론방법
Case-based Reasoning (CBR)  
  
(k-nn)알고리즘을 이용하여 검색된 ___유사 과거 사례들을 조합___ 하여 새로운 문제의 정답을 도출  

기존의 사건들을 일정 규칙을 통해 case base라는 공간에 보관  

새로운 사건 발생시 case base에 물어보고 어떻게 문제를 해결해야 할지 결정
case누적시 유용성 증가  


6. 로지스틱 회귀분석

집단 분류 기준 찾아내는 방식 회귀방식으로 분류문제를 푼다.

## 4. 모형평가
분석 기법들을 사용하고 나면 이에 따른 결과들의 정확도를 판단할 수 있어야 한다.  


__모형평가__ : 모델의 유용성을 판단, 상이한 모형들을 서로 비교하여 평가
※결과변수를 알고있는 지도학습(Supervised learning)환경에서 사용  
  
- 분류 모형의 성과 평가 : 분류 정확도와 이익도표, ROC 도표      
- 예측 모형의 성과 평가 : 평균절대오차, 평균 오차, 평균 절대 비율 오차, 평균 제곱 오차의 제곱근, 제곱오차의 총합계    

>- 분류 정확도_ 분류 결과의 정확성을 평가로 결과선택     
        정분류율, 오분류율, 민감도, 특이도에 따라 나눔  
>- 이익 도표_ 모형의 결과를 바탕으로 분류 정확도를 측정   
        사례를 적게 선택하고 상대적으로 높은 응답자의 비율을 찾음으로써 가장 좋은 성과를 보이는 부분을 효과적으로 찾고자 할 때 유용 ___(이게 무슨소리인지..)___  
>- ROC 도표_ 민감도와 특이도 이용 분류 정확성 검증  
        민감도를 집단 정확 판별 지표   
        특이도는 중요하지 않은 집단을 정확 판별   
        민감도_특이도 trade-off관계
